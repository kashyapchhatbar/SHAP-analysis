```python
dataset = "data/SET1A_ZC3H4_Hughes_et_al"
best_params = find_best_params(dataset)

r2_train = []
r2_test = []
for n, current_params in best_params:
    a, b, model = model_data(current_params, dataset)
    r2_train.append(a)
    r2_test.append(b)
hughes_metrics = pd.DataFrame([r2_train, r2_test, ["Hughes et al."]*5]).T

dataset = "data/INTS11_Wang_et_al"
best_params = find_best_params(dataset)

r2_train = []
r2_test = []
for n, current_params in best_params:
    a, b, model = model_data(current_params, dataset)
    r2_train.append(a)
    r2_test.append(b)
wang_metrics = pd.DataFrame([r2_train, r2_test, ["Wang et al."]*5]).T

dataset = "data/PRC1_PRC2_Dobrinic_et_al"
best_params = find_best_params(dataset)

r2_train = []
r2_test = []
for n, current_params in best_params:
    a, b, model = model_data(current_params, dataset)
    r2_train.append(a)
    r2_test.append(b)
dobrinic_metrics = pd.DataFrame([r2_train, r2_test, ["Dobrinić et al."]*5]).T
```

## Figure 1B

```python
df = pd.concat([hughes_metrics, wang_metrics, dobrinic_metrics])
df.columns = ["train", "test", "Dataset"]

fig, ax = plt.subplots(figsize=(6.75,2))
sns.stripplot(data=pd.melt(df, id_vars="Dataset"),
              x="value", y="Dataset", hue="variable", ax=ax, palette=sns.color_palette(["grey", "black"]),
             edgecolor="white", linewidth=0.5, jitter=0.2, size=10)
ax.set_xlim(0, 1)
ax.grid(True, which="both")
ax.set_xlabel("Coefficient of variation (R²)")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_ylabel("")
fig.tight_layout()
```
